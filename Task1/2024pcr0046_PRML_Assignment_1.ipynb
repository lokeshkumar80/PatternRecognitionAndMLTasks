```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

#**Question 1 : Bayes Classifier**



```python
# Given parameters
mu1, sigma1, P_w1 = 5, 1.5, 0.6
mu2, sigma2, P_w2 = 8, 2.0, 0.4

# Define the range of x values
x_values = np.linspace(0, 15, 100)

# Compute class-conditional probabilities (Gaussian PDFs)
p_x_w1 = norm.pdf(x_values, mu1, sigma1)
p_x_w2 = norm.pdf(x_values, mu2, sigma2)

# Compute P(x)
P_x = (p_x_w1 * P_w1) + (p_x_w2 * P_w2)
```

# Task 1 : Posterior Probabilities

```python
# Compute posterior probabilities using Bayes' theorem
P_w1_x = (p_x_w1 * P_w1) / P_x
P_w2_x = (p_x_w2 * P_w2) / P_x
```

```python
#Print Posterior Probabilities
print(P_w1_x)
```

```python
#Print Posterior Probabilities
print(P_w2_x)
```

# Task 2 : Plot the Graph

```python
# Find the decision boundary where P(w1|x) = P(w2|x)
decision_boundary = x_values[np.argmin(np.abs(P_w1_x - P_w2_x))]

# Plot the results
plt.figure(figsize=(8, 5))
plt.plot(x_values, P_w1_x, label=r'$P(\omega_1 | x)$', color='blue')
plt.plot(x_values, P_w2_x, label=r'$P(\omega_2 | x)$', color='red')
plt.axvline(decision_boundary, color='black', linestyle='--', label='Decision Boundary')
plt.xlabel("Feature x")
plt.ylabel("Posterior Probability")
plt.title("Posterior Probabilities P(ω1|x) and P(ω2|x)")
plt.legend()
plt.grid()
plt.show()
```

#Task 3 : Decision Rule

```python
# Find the decision boundary where P(w1|x) = P(w2|x)
decision_boundary = x_values[np.argmin(np.abs(P_w1_x - P_w2_x))]

# Plot the results
plt.figure(figsize=(8, 5))
plt.plot(x_values, P_w1_x, label=r'$P(\omega_1 | x)$', color='blue')
plt.plot(x_values, P_w2_x, label=r'$P(\omega_2 | x)$', color='red')
plt.axvline(decision_boundary, color='black', linestyle='--', label='Decision Boundary')

# Shade decision regions
plt.fill_between(x_values, 0, 1, where=(P_w1_x > P_w2_x), color='blue', alpha=0.1)
plt.fill_between(x_values, 0, 1, where=(P_w2_x > P_w1_x), color='red', alpha=0.1)

plt.xlabel("Feature x")
plt.ylabel("Posterior Probability")
plt.title("Posterior Probabilities P(ω1|x) and P(ω2|x) with Decision Regions")
plt.legend()
plt.grid()
plt.show()
```

#Task 5 : Discuss the misclassification rate if the decision boundary is shifted slightly.

```python
# Compute misclassification rate for slightly shifted boundary
def misclassification_rate(shift):
    new_boundary = decision_boundary + shift
    P_error_w1 = P_w1 * norm.cdf(new_boundary, mu1, sigma1)
    P_error_w2 = P_w2 * (1 - norm.cdf(new_boundary, mu2, sigma2))
    return P_error_w1 + P_error_w2

shifts = np.linspace(0, 10, 100)
misclassification_rates = [misclassification_rate(shift) for shift in shifts]

plt.figure(figsize=(8, 5))
plt.plot(shifts, misclassification_rates, marker='o', linestyle='-', color='purple')
plt.axvline(decision_boundary, color='black', linestyle='--', label='Original Boundary')
plt.xlabel('Shift in Decision Boundary')
plt.ylabel('Misclassification Rate')
plt.legend()
plt.title('Effect of Decision Boundary Shift on Misclassification Rate')
plt.show()
```

#**Question 3 :** **Bayesian Decision Theory with Loss Function**


```python
# Given values
P_w1 = 0.4
P_w2 = 0.6
mean1, std1 = 550, 20
mean2, std2 = 700, 25
loss_11, loss_12, loss_21, loss_22 = 0, 2, 5, 0

x_values = np.linspace(450, 800, 1000)
```

Task 1: Compute the posterior probabilities P(ω1|x) and P(ω2|x) and          Task 2: Compute the expected loss for both actions R(α1|x) and R(α2|x)

```python
def class_conditional_pdf(x, mean, std):
    return norm.pdf(x, mean, std)

def posterior_probability(x, prior, mean, std, evidence):
    return (class_conditional_pdf(x, mean, std) * prior) / evidence

def compute_expected_loss(x_values, P_w1, P_w2, mean1, std1, mean2, std2, loss_11, loss_12, loss_21, loss_22):
    px_w1 = class_conditional_pdf(x_values, mean1, std1)
    px_w2 = class_conditional_pdf(x_values, mean2, std2)
    evidence = P_w1 * px_w1 + P_w2 * px_w2
    P_w1_x = posterior_probability(x_values, P_w1, mean1, std1, evidence)
    P_w2_x = 1 - P_w1_x

    R_alpha1_x = loss_11 * P_w1_x + loss_12 * P_w2_x
    R_alpha2_x = loss_21 * P_w1_x + loss_22 * P_w2_x

    decision_boundary_index = np.argmin(np.abs(R_alpha1_x - R_alpha2_x))
    return x_values[decision_boundary_index], R_alpha1_x, R_alpha2_x
```


#Task 3 :  Find the decision boundary where the expected loss is equal for both actions

```python
decision_boundary, R_alpha1_x, R_alpha2_x = compute_expected_loss(x_values, P_w1, P_w2, mean1, std1, mean2, std2, loss_11, loss_12, loss_21, loss_22)
```

#Task 4 : Plot the expected loss functions R(α1|x) and R(α2|x)

```python
plt.figure(figsize=(10, 5))
plt.plot(x_values, R_alpha1_x, label='R(α1|x)', color='red')
plt.plot(x_values, R_alpha2_x, label='R(α2|x)', color='blue')
plt.ylabel('Expected Loss')
plt.legend()
plt.title('Expected Loss Functions')
plt.show()



```

#Task 5 : Discuss how the decision boundary changes if the prior probability of high-risk applicants increases

```python
plt.figure(figsize=(10, 5))
plt.plot(x_values, R_alpha1_x, label='R(α1|x)', color='red')
plt.plot(x_values, R_alpha2_x, label='R(α2|x)', color='blue')
plt.axvline(decision_boundary, color='black', linestyle='--', label='Decision Boundary')
plt.xlabel('Credit Score x')
plt.ylabel('Expected Loss')
plt.legend()
plt.title('Expected Loss Functions')
plt.show()

print(f"Decision boundary is at credit score = {decision_boundary:.2f}")

# Analyze effect of increasing prior probability of high-risk applicants
P_w1_new = 0.5  # Increased probability of high risk applicants
P_w2_new = 1 - P_w1_new

decision_boundary_new, _, _ = compute_expected_loss(x_values, P_w1_new, P_w2_new, mean1, std1, mean2, std2, loss_11, loss_12, loss_21, loss_22)
print(f"New decision boundary with increased high-risk prior: {decision_boundary_new:.2f}")
```

#**Question 4: Risk Minimization with Three Classes**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Given values
priors = [0.2, 0.5, 0.3]
means = [2, 5, 10]
stds = [0.5, 1.0, 1.5]
loss_matrix = np.array([[0, 5, 10], [3, 0, 7], [8, 6, 0]])
x_values = np.linspace(0, 15, 1000)
```

#Task 1:  Compute the posterior probabilities P(ω1|x), P(ω2|x), and P(ω3|x) & Task 2:  Compute the expected loss for each action.

```python
def class_conditional_pdf(x, mean, std):
    return norm.pdf(x, mean, std)

def posterior_probability(x, prior, mean, std, evidence):
    return (class_conditional_pdf(x, mean, std) * prior) / evidence

def compute_expected_loss(x_values, priors, means, stds, loss_matrix):
    class_pdfs = [class_conditional_pdf(x_values, mean, std) for mean, std in zip(means, stds)]
    evidence = sum(prior * pdf for prior, pdf in zip(priors, class_pdfs))
    posteriors = [(prior * pdf) / evidence for prior, pdf in zip(priors, class_pdfs)]

    expected_losses = [sum(loss_matrix[i, j] * posteriors[j] for j in range(len(priors))) for i in range(len(priors))]
    expected_losses = np.array(expected_losses)
    optimal_decision_indices = np.argmin(expected_losses, axis=0)

    return expected_losses, optimal_decision_indices
```

# Task 3: Find the optimal decision boundary that minimizes the risk

```python
expected_losses, optimal_decision_indices = compute_expected_loss(x_values, priors, means, stds, loss_matrix)
```

# Task 4 : Plot the decision boundaries and label regions for different classes.

```python
# Plot decision boundaries
plt.figure(figsize=(10, 5))
plt.plot(x_values, expected_losses[0], label='R(α1|x) - Pedestrian', color='red')
plt.plot(x_values, expected_losses[1], label='R(α2|x) - Vehicle', color='blue')
plt.plot(x_values, expected_losses[2], label='R(α3|x) - Empty Road', color='green')

# for i in range(1, len(x_values)):
#     if optimal_decision_indices[i] != optimal_decision_indices[i - 1]:
#         plt.axvline(x_values[i], color='black', linestyle='--')

plt.xlabel('Distance from LiDAR Sensor (x)')
plt.ylabel('Expected Loss')
plt.legend()
# plt.title('Expected Loss and Decision Boundaries')
plt.show()
```

#Task 5 : Discuss the effect of increasing the penalty for misclassifying a pedestrian.

```python
# Plot decision boundaries
plt.figure(figsize=(10, 5))
plt.plot(x_values, expected_losses[0], label='R(α1|x) - Pedestrian', color='red')
plt.plot(x_values, expected_losses[1], label='R(α2|x) - Vehicle', color='blue')
plt.plot(x_values, expected_losses[2], label='R(α3|x) - Empty Road', color='green')

for i in range(1, len(x_values)):
    if optimal_decision_indices[i] != optimal_decision_indices[i - 1]:
        plt.axvline(x_values[i], color='black', linestyle='--')

plt.xlabel('Distance from LiDAR Sensor (x)')
plt.ylabel('Expected Loss')
plt.legend()
plt.title('Expected Loss and Decision Boundaries')
plt.show()

# Effect of increasing pedestrian misclassification penalty
loss_matrix[1, 0] = 10  # Increased penalty for misclassifying pedestrian as vehicle
expected_losses_new, optimal_decision_indices_new = compute_expected_loss(x_values, priors, means, stds, loss_matrix)

plt.figure(figsize=(10, 5))
plt.plot(x_values, expected_losses_new[0], label='R(α1|x) - Pedestrian', color='red')
plt.plot(x_values, expected_losses_new[1], label='R(α2|x) - Vehicle', color='blue')
plt.plot(x_values, expected_losses_new[2], label='R(α3|x) - Empty Road', color='green')

for i in range(1, len(x_values)):
    if optimal_decision_indices_new[i] != optimal_decision_indices_new[i - 1]:
        plt.axvline(x_values[i], color='black', linestyle='--')

plt.xlabel('Distance from LiDAR Sensor (x)')
plt.ylabel('Expected Loss')
plt.legend()
plt.title('Effect of Increased Pedestrian Misclassification Penalty')
plt.show()
```
